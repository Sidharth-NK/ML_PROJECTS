```python
teams = pd.read_csv("/teams.csv")
```

```python
teams  = teams[["team","country","year","athletes","age","prev_medals","medals"]]
teams
```

```python
#to check and remove the missiing values in the dataset. This is DATA CLEANING process
teams[teams.isnull().any(axis=1)]
```

```python
teams = teams.dropna() # dropna() removes all the null value/ missing value data in the dataset
teams
```

```python
# creating test data and traning data. make sure you use the old past data not the future data inorder to avoid errors

train = teams[teams["year"]<2012].copy() # copies the data before 2012 to the training set
test = teams[teams["year"]>=2012].copy() # copies the data of 2012 and after to the test set
```

```python
train.shape # size of training data set (row,columns)
```

```python
test.shape
 # Here the training data and the test data are in 80% 20% ratio. which is the ideal/required ratio
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target = "medals"
```

```python
reg.fit(train[predictors],train[target]) # this is to fit the data in the linear model
```

```python
prediction = reg.predict(test[predictors]) # this is to predict the data in the linear model
# here only predictors is given so as to check whether the model is correctly predicting.
# prediction -> this will print numpy arrays without rounded and with negative values.
test["predictions"] = prediction
test.loc[test["predictions"]<0,"predictions"] = 0 # this will locate all the negative values and replace it with zeros
```

```python
# values are rounded to the nearest whole number, to make it more readable and understandable
test["predictions"] = test["predictions"].round()
test
```

```python
from sklearn.metrics import mean_absolute_error as mae
error =  mae(test["medals"],test["predictions"])
error
#this means that on avg we were 3.3 medals of how many medals a team actually won in the olympics
```
